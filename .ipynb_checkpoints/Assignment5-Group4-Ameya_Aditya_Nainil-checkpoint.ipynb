## Resizing the Images

Now lets resize every picture first.

We are resizing every picture to have a height of 4800 and a corresponding width. 

For this task, we define a function `shape_res_height` where we take in the dimensions of the image `size` and required height `max_height` as parameters, then scale the width as per the required height and return the new dimensions

```python
def shape_res_height(size, max_height):
    # Take the original width and height
    size_width = size[0]
    size_height = size[1]
    
    # Rescale the width to match the aspect ratio of the new height
    new_width = round(size_width*max_height/size_height)
    
    # Form an array of the new dimensions
    new_size = [new_width, max_height]
    
    # Return the new dimensions
    return new_size
```

As specified earlier, we would now resize all the images to have a height of 4800, so that the images could be easily stitched horizontally since their heights would be the same. 

In the below code block, we index through the entire dataframe `data` and resize each of the images using the new dimensions obtained from the `shape_res_height` function. 

We then finally save all the resized images in the `resized` folder nested inside our `comicPhotos` folder.

```python
height = 4800
for ind in data.index:
    img = pili.open('comicPhotos/' + data['Photo Name'][ind] +'.jpg')
    print(data['Photo Name'][ind] + ' resized')
    size = img.size
    print('old size' + str(img.size))
    new_dims = shape_res_height(size, height)

    img = img.resize(new_dims).convert('RGB')
    print('new size' + str(img.size))
    
    img.save('comicPhotos/resized/' + data['Photo Name'][ind] +'.jpg')    
```

## Cartoonizing and bordering the images

First we define a function `imgcompress_mem`, to compress the photos. 

Here we take the **image path** and a resize parameter **k** as the parameters and then resize the dimensions of the image on the basis of the parameter **k**. 

Finally we return back the image without actually saving it

```python
def imgcompress_mem(path_in, k):
    img = cv2.imread(path_in, cv2.IMREAD_UNCHANGED)

    # set the ratio of resized image
    width = int((img.shape[1])/k)
    height = int((img.shape[0])/k)

    # resize the image by resize() function of openCV library
    return cv2.resize(img, (width, height), interpolation=cv2.INTER_AREA)
```

We also define a new function `cartoonizebl_mem` to cartoonize each of our image. 

Here we first take in the image and pass it to our image compression function `imgcompress_mem` defined earlier. 

We then take in the `blur` amount and `line` size as the paramters. 

We then first convert our compressed image to grayscale image using `cv2.cvtColor(imgc, cv2.COLOR_BGR2GRAY)` and then obtain a blurred version of our grayscaled image using `cv2.medianBlur(gray, blur_value)`. 

We then apply the `cv2.adaptiveThreshold` function which sets each of the pixel value to maximum or minimum depending upon the threshold. We make use of `ADAPTIVE_THRESH_GAUSSIAN_C` as the threshold parameter so as to have the threshold value as the gaussian-weighted sum, which in turn gives better edges in our result.

We then finally convert this image to `RGB` and take the `bitwise AND` of our original and edged image to get the cartoonized version of the image

```python
def cartoonizebl_mem(path_in, k, blur, line):
    # Compress the source image
    imgc = imgcompress_mem(path_in, k)

    # Initialize the line width and blur amount
    line_size = line
    blur_value = blur

    # Convert the RGB image to GRAY image
    gray = cv2.cvtColor(imgc, cv2.COLOR_BGR2GRAY)
    # Blur the GRAY image upto the required amount
    gray_blur = cv2.medianBlur(gray, blur_value)
    
    # Applying the adaptiveThreshold function which sets the pixel value to 0 if it is smaller than the threshold, otherwise sets to a maximum value.
    # we also select ADAPTIVE_THRESH_GAUSSIAN_C as the threshold paramter. The threshold value is a gaussian-weighted sum of the neighbourhood values minus the constant C.
    bigedges = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, line_size, blur_value)
    
    # Converting BGR to RGB
    bigedges_pil = cv2.cvtColor(bigedges, cv2.COLOR_BGR2RGB) 
    
    #return the bitwise AND of the original and cartoonized image
    return cv2.bitwise_and(imgc, imgc, mask=bigedges)
```

Now lets define a function `add_caption` that will first add caption to our image and then cartoonize it using the `cartoonizebl_mem` function we defined above.

In this function, we take the **image name, caption, caption language, caption alignment, caption font and caption font size** as the input parameters

We first define the tint, transparency and opacity of our caption box. We then open our image and draw an overlay for our caption text and also dynamically set the font and font size.

We then get the default caption size and also the number of lines in our caption. We then further initialize the coordinates from where we want our caption to begin. The X coordinate is set to a default 150 but the Y coordinate is calculated on the basis of the number of lines existing in the caption

We then create and draw a rectangle of required opacity and color for our caption and then insert our caption text inside this rectangle. We then alpha composite this overlay and our original image to create the image with caption

We then save this image with captions in the `withCaptionLanguage` folder created inside the `comicPhotos` folder.

Lastly, we cartoonize this image using our `cartoonizebl_mem` function and then remove the alpha component and return the result image 

```python
def add_caption(caption, caption_language, caption_alignment, caption_font, caption_font_size, image_name):
    TINT_COLOR = (0, 0, 0)  # Black
    TRANSPARENCY = 0.5  # Degree of transparency, 0-100%
    OPACITY = int(255 * TRANSPARENCY)

    # Load the image
    img = pili.open(r'comicPhotos/resized/'+image_name+'.jpg').convert('RGBA')
    
    # Create the overlay for the caption text
    overlay = pili.new('RGBA', img.size, TINT_COLOR+(0,))
    draw = pild.Draw(overlay)
    
    # Get dynamic font and font size
    font = pilf.truetype(caption_font + ".ttf", caption_font_size)

    # Get the caption size
    w, h = font.getsize(caption)
    
    # Split the caption at \n to get the number of lines
    num_lines = len(caption.split('\n'))
    
    # Initialize the x and y coordinates from where the caption begins
    x, y = 150, img.height - (num_lines)*h - 150
    
    # Draw a rectangle of required color and opacity around the caption
    draw.rectangle((0.9*x, 0.99*y, img.width-(0.9*x), 1.01*(y + (num_lines)*h + 25)), fill=TINT_COLOR+(OPACITY,))
    
    # Insert the caption text in the rectangle
    draw.text((1.5*x, y), caption, fill=(255, 255, 8), font=font, align=caption_alignment, spacing = 40)

    # Alpha composite these two images together to obtain the desired result.
    img = pili.alpha_composite(img, overlay)
    
    # Remove alpha for saving in jpg format.
    img = img.convert("RGB") 

    # Save the image with caption in 'withCaptions' folder
    img.save('comicPhotos/withCaption'+ caption_language +'/'+image_name+'.jpg')
    
    # Cartoonize and return the image
    cblimg = cartoonizebl_mem('comicPhotos/withCaption'+ caption_language +'/'+image_name+'.jpg', 1, 5, 101)
    cblimg_pil = cv2.cvtColor(cblimg, cv2.COLOR_BGR2RGBA) # Converting BGR to RGB
    return cblimg_pil
```

Now, lets cartoonize each image and add borders!

First set the height of the images along with the border as 5000 in `height`. 
We then index through the entire dataframe and for each image pass the required paramters to the `add_caption` function.

We then add borders to each of these images, a border height equal to (5000 - 4800)/2 and a border width of 200.

Then, we convert these images with borders to `RGB` and save them in `withCaptionEnglish` and `withCaptionMarathi` folder respectively for images with English and Marathi captions.

Finally we display if the processing of the respective image with respective caption language was completed

```python
height = 5000

for ind in data.index:
    img_e = add_caption(data['English dialogs'][ind], 'English', 'left', 'OpenSans', data['Font Size'][ind], data['Photo Name'][ind])
    img_m = add_caption(data['Marathi dialogs'][ind], 'Marathi', 'left', 'NirmalaS', data['Font Size'][ind], data['Photo Name'][ind])
    
    new_height = round((height - img_e.shape[0])/2)

    fimg_e = piliops.expand(pili.fromarray(np.uint8(img_e)).convert('RGB'), border=(200,new_height),fill=(128,128,0)).save(r'comicPhotos/withCaptionEnglish/'+data['Photo Name'][ind]+'.jpg')
    print(data['Photo Name'][ind] + ' in English - Completed')
    fimg_m = piliops.expand(pili.fromarray(np.uint8(img_m)).convert('RGB'), border=(200,new_height),fill=(128,128,0)).save(r'comicPhotos/withCaptionMarathi/'+data['Photo Name'][ind]+'.jpg')
    print(data['Photo Name'][ind] + ' in Marathi - Completed')
```

## Stacking images horizontally

We now define a function to stack the cartoonized images horizontally.

In this function, we take in a dataframe, the language, page number and strip number as the input parameters.

We then create a list of possible images from the dataframe and load them. We then stack them horizontally using `np.hstack()` and the convert the image from array format to image using `pili.fromarray`. 

We then finally save the image in the `withCaptionEnglish` or `withCaptionMarathi` folder respectively for the languages with the naming convention being `pageNumber_stripNumber.jpg`

We finally display if the processing of a strip on a particular page in a particular language was completed

```python
# Here we make a function to stack images horizontally that were just created above
# We take the dataframe containing the data of the necessary images along with the language, page number and strip number as the paramters

height = 5000
def image_stack_horizontal(data, language, page_no, strip_no):
    # First we create a list of all images from the dataframe
    list_images = ['comicPhotos/withCaption'+language+'/'+data['Photo Name'][ind]+'.jpg' for ind in data.index]
    
    # We then open all the images using pillow
    imgs = [pili.open(i) for i in list_images]
    
    # Since the images have the same vertical size, we just stack them horizontally by stitching all of them
    imgs_comb = np.hstack((np.asarray(i) for i in imgs))

    # We then convert the image from the array format and save the combined image
    imgs_comb = pili.fromarray(imgs_comb)
    imgs_comb.save('comicPhotos/withCaption'+language+'/'+str(page_no) + '_' + str(strip_no)+'.jpg')  

    # Finally we display which image was just recently generated
    print('Language: ' + language + ', Page ' + str(page_no) + ', Strip ' + str(strip_no) + ' Done')
```

Here, we first slice the dataframe according to the page number and then extract the unique strip numbers from the sliced dataframe.

We then sort the sliced dataframe according to the `Column no` colomn so that all the images would be in the required sequence in the sliced dataframe.

We then pass this sliced dataframe to the `image_stack_horizontal` function along with the required parameters which then creates and stores the required horizontally stacked images in the above mentioned folders

```python
for page_no in data['Page no'].unique():
    data_pageno = data.loc[data['Page no'] == page_no]
    for strip in data_pageno['Strip'].unique():
        data_strip = data_pageno.loc[data['Strip'] == strip]
        data_strip = data_strip.sort_values('Column no',ignore_index=True)
        image_stack_horizontal(data_strip, 'English', data_strip['Page no'][0], data_strip['Strip'][0])
        image_stack_horizontal(data_strip, 'Marathi', data_strip['Page no'][0], data_strip['Strip'][0])
```

## Resizing the horizontally stacked images and stacking them vertically

Now lets resize every horizontally stacked picture. 

For this we define a function `shape_res_width` where we take in the dimensions `size` of the stacked image, then scale the height as per the `max_width` that we require and return the new dimensions

```python
def shape_res_width(size, max_width):
    # Take the original width and height
    size_width = size[0]
    size_height = size[1]
    
    # Rescale the height to match that of the new height
    new_height = round(size_height*max_width/size_width)
    
    # Form an array of the new dimensions
    new_size = [max_width, new_height]
    
    # Return the new dimensions
    return new_size
```

Lets, first define the page dimensions 

```python
page_width = 2480 
page_height = 3508 
```

Now lets define a function `generate_all_pages` which would generate all the pages provided we pass a language as an input parameter.

The function takes in the language and generates a list of all the strips possible for every page and loads the required images.

It then takes in the maximum width from the list of images and resizes all the other images accordingly.

It then vertically stacks the images using the `np.vstack()` function and converts it to an image using the `fromarray` function.

It then resizes the vertically stacked image according to page dimensions and adds a border to every page.

Finally this vertically stacked image along with the borders is saved and the operation completion is signified

```python
def generate_all_pages(language):
    # We first iterate through all the pages 
    for page_no in data['Page no'].unique():
        data_page_no = data.loc[data['Page no'] == page_no]
        
        # Sort the dataframe according to the Strip numbers and create a list of all possible lists on the page
        data_page_no = data_page_no.sort_values('Strip',ignore_index=True)
        list_images_page = ['comicPhotos/withCaption'+ language + '/' + str(i) + '_' + str(j) + '.jpg' for i in data_page_no['Page no'].unique() for j in data_page_no['Strip'].unique()]

        # Load the required images
        imgs = [pili.open(i) for i in list_images_page]

        # Pick the image which has the maximum width and keep that selected width in max_width
        max_width = sorted(i.size[0] for i in imgs)[0]

        # Then resize the images using the shape_res_width function and vertically stack them
        imgs_comb = np.vstack( (np.asarray( i.resize(shape_res_width(i.size,max_width))) for i in imgs ) )
        imgs_comb = pili.fromarray(imgs_comb)

        ## Make the vertically stitched image's dimensions according to our page dimensions
        width, height = imgs_comb.size
        width_ratio = (width/page_width)
        height_ratio = (height/page_height)
        main_ratio = max(width_ratio, height_ratio)

        # Resize the image to fit the page dimensions
        imgs_comb = imgs_comb.resize((int(round(width/main_ratio)), int(round(height/main_ratio))))

        # Add a border to the resized image and save the image
        fimg_e = piliops.expand(imgs_comb, border=(round((page_width - imgs_comb.size[0])/2),round((page_height - imgs_comb.size[1])/2)),fill=(128,128,0)).save('comicPhotos/withCaption'+language+'/page' +  str(page_no) +'.jpg')

        print('Language: '+ language +' Page ' + str(page_no) + ' Completed')
```

We now generate all the pages for both the English and Marathi story book by calling the `generate_all_pages` function

```python
generate_all_pages('English')
generate_all_pages('Marathi')
```

## Generating the PDF

We now finally define the `make_pdf` which generates a list of image names from the dataframe and then loads the images.

It only takes the language as an input parameter.

It then creates a pdf and appends every image as a new page in the pdf with the page dimensions equal to the image dimensions.

It finally outputs the result pdf with the naming convention `story_book_language` and is stored in the `withCaptionLanguage` folders for the respective languages

```python
def make_pdf(language):
    # Generate a list of all images
    imagelist = ['comicPhotos/withCaption'+language + '/' + 'page' +  str(i) + '.jpg' for i in data['Page no'].unique()]
   
    # Load the images
    imgs    = [pili.open(i) for i in imagelist]

    # Start generating the pdf
    pdf = FPDF(unit = "pt", format = [page_width, page_height])
    
    # Add cover page
    img = pili.open(language+' Front.jpg')
    width, height = img.size
        
    pdf.add_page()
    pdf.image(language + ' Front.jpg', 0, 0, width, height)

    # Iterate over the imagelist and append all the images as subsequent pages
    for image in imagelist:
        img = pili.open(image)
        width, height = img.size
        
        pdf.add_page()
        pdf.image(image, 0, 0, width, height)
        
    # Save the generated pdf in the respective folder
    pdf.output('comicPhotos/withCaption'+language + '/' + 'story_book_' + language + '.pdf', "F")
    
    # Signify that the pdfs are generated
    print(language + ' Pdf Done!!!')
```

Calling the `make_pdf` function to generate the pdfs for both languages

```python
make_pdf('English')
make_pdf('Marathi')
```
